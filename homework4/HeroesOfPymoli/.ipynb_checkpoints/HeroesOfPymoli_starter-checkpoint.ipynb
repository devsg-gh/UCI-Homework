```python
# Dependencies and Setup
import pandas as pd

# File to Load 
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)

purchase_data.head()
```

## Player Count

* Display the total number of players


```python
#count the number of players using count
player_count = purchase_data["SN"].nunique()
player_count
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# calcualte, unique items, mean of prices, purchase counts, sum of price
noui =purchase_data["Item ID"].nunique()
avgp =purchase_data["Price"].mean()
tnop =purchase_data["Purchase ID"].count()
trev =purchase_data["Price"].sum()
# create a dataframe for analysis
pur_analysis = pd.DataFrame([{"Number of Unique Items": noui, 
                              "Average Price": avgp,
                              "Number of Purchases": tnop, 
                               "Total Revenue": trev}])

#Formating the Price 
pur_analysis['Average Price'] = pur_analysis['Average Price'].map('${:,.2f}'.format)
pur_analysis['Total Revenue'] = pur_analysis['Total Revenue'].map('${:,.2f}'.format)
pur_analysis
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#calclate the details by dropping duplicate names
result_df = purchase_data.drop_duplicates(subset=['SN'], keep='first')
# to count the total number of unique players, male, female and others
count_sn = result_df["SN"].count()
uniq_gender =purchase_data["Gender"].unique()
male_count =len(result_df[result_df["Gender"] == "Male"])
female_count =len(result_df[result_df["Gender"] == "Female"])
other_count =len(result_df[result_df["Gender"] == "Other / Non-Disclosed"])
#calculate the % of the male, female and others
per_male = (round((male_count/count_sn)*100,2)).astype(str) + '%'
per_female = (round((female_count/count_sn)*100,2)).astype(str) + '%'
per_othr = (round((other_count/count_sn)*100,2)).astype(str) + '%'
#create dataframe for display
gender_demo = pd.DataFrame({
            "Total Count": [male_count, female_count, other_count], 
            "Percentage of Players": [per_male, per_female, per_othr],
            
            "": ["Male", "Female", "Other/Non-Disclosed"]
                       })

gender_demo_index = gender_demo.set_index("")
gender_demo_index
#count_sn
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#calculate the sum of price for Male, female and others
pur_cnt_m = purchase_data.loc[purchase_data["Gender"] == "Male", "Price"].sum()
pur_cnt_f = purchase_data.loc[purchase_data["Gender"] == "Female", "Price"].sum()
pur_cnt_o = purchase_data.loc[purchase_data["Gender"] == "Other / Non-Disclosed", "Price"].sum()

#calculate the count for Male, female and others
male_count_nu =len(purchase_data[purchase_data["Gender"] == "Male"])
female_count_nu =len(purchase_data[purchase_data["Gender"] == "Female"])
other_count_nu =len(purchase_data[purchase_data["Gender"] == "Other / Non-Disclosed"])

#create dataframe with details
purchase_analysis = pd.DataFrame({
            "Purchase Count": [female_count_nu, male_count_nu, other_count_nu], 
            "Average Purchase Price": [round(pur_cnt_f/female_count_nu,2), round(pur_cnt_m/male_count_nu,2), round(pur_cnt_o/other_count_nu,2)],
            "Total Purchase Value": [pur_cnt_f, pur_cnt_m, pur_cnt_o],
            "Avg Total Purchased per Person": [(round(pur_cnt_f/female_count,2)), 
                                               (round(pur_cnt_m/male_count,2)),
                                               (round(pur_cnt_o/other_count,2))],
            "": ["Male", "Female", "Other/Non-Disclosed"]
                      })
#format data 
purchase_analysis['Average Purchase Price'] = purchase_analysis['Average Purchase Price'].map('${:,.2f}'.format)
purchase_analysis['Total Purchase Value'] = purchase_analysis['Total Purchase Value'].map('${:,.2f}'.format)
purchase_analysis['Avg Total Purchased per Person'] = purchase_analysis['Avg Total Purchased per Person'].map('${:,.2f}'.format)

gender_demo_index = purchase_analysis.set_index("")
gender_demo_index

```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Create the bins in which Data will be held
# Bins   
bins = [0, 9, 14, 19, 24, 29, 34, 39, 49]

# Create the names for the bins
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

age_df=result_df.copy(deep=True)

#add age group in the df to display group names

age_df["age_group"] = pd.cut(age_df["Age"], bins, labels=group_names)

#calculate count by age group
numbers_age_group= age_df["age_group"].value_counts()

#calculate % by age group
percentage_age_group = round((numbers_age_group/count_sn)*100,2)

age_demo =pd.DataFrame({"Total Count":numbers_age_group,
                        "Percentage":percentage_age_group})

age_demo['Percentage'] = age_demo['Percentage'].map('{:,.2f}%'.format)


age_demo =age_demo.sort_index(0, ascending= True)
age_demo
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Create the bins in which Data will be held
# Bins
bins = [1, 9, 14, 19, 24, 29, 34, 39, 100]
#create the value for bins
group_names = ["<10","10-14","15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#add age group field to the current dataframe
purchase_data["age_group"] = pd.cut(purchase_data["Age"], bins, labels=group_names)

#Create new dataframe
purchase_data_age_group_df = pd.DataFrame(purchase_data, columns = ["SN","age_group", "Price"])

#create dataset for purchase count by age
purchase_count = purchase_data_age_group_df.groupby(["age_group"])["Price"].count()

#create dataset for Average Price in age group
purchase_price_by_age_group = purchase_data_age_group_df.groupby(["age_group"])["Price"].sum()
avg_purchase_price =  round((purchase_price_by_age_group/purchase_count),2) 

player_count_by_age_group = purchase_data_age_group_df.groupby(["age_group"])["SN"].count()
#Format the price data and add $ sign
avg_total_purchase_per_person = purchase_price_by_age_group/numbers_age_group

#Create Dataframe
purchasing_analysis = pd.DataFrame({
    "Purchase Count": purchase_count,
    "Average Purchase Price": avg_purchase_price.map("${:.2f}".format),
    "Total Purchase Value": purchase_price_by_age_group.map("${:.2f}".format),
    "Avg Total Purchase per Person": avg_total_purchase_per_person.map("${:.2f}".format)
    
})

purchasing_analysis =purchasing_analysis.sort_index(0, ascending= True)
purchasing_analysis
```

## Top Spenders

* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Get data for count by Name and Item
purch_count = purchase_data.groupby(["SN"])["Item ID"].count()
#Get data for total purchase by Name and Price
total_purchase_value = purchase_data.groupby(["SN"])["Price"].sum()
#Get data for average price
average_purchase_price =  round((total_purchase_value/purch_count),2) 

#Create Dataframe for Analysis
top_spend = pd.DataFrame({
    "Purchase Count": purch_count,
    "Average Purchase Price": average_purchase_price,
    "Total Purchase Value": total_purchase_value
    
})
#format the data
top_spend["Average Purchase Price"] = top_spend['Average Purchase Price'].map("${:.2f}".format)
#top_spend["Total Purchase Value"]= top_spend['Total Purchase Value'].astype(float).map("${:.2f}".format)

top_spend =top_spend.sort_values("Total Purchase Value", ascending= False)
top_spend.head()


```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Get data for purchase id count, price sum and mean by  Item id and Item name
popular_count = purchase_data.groupby(["Item ID","Item Name"])["Purchase ID"].count()
total_purchase_val = purchase_data.groupby(["Item ID","Item Name"])["Price"].sum()
popular_item_price = purchase_data.groupby(["Item ID","Item Name"])["Price"].mean()

#Create Dataframe for Analysis
popular_item = pd.DataFrame({
    "Purchase Count": popular_count,
    "Item Price": popular_item_price,
    "Total Purchase Value": total_purchase_val
})
#format the data
popular_item["Item Price"] = popular_item['Item Price'].map("${:.2f}".format)
#popular_item["Total Purchase Value"] = popular_item['Total Purchase Value'].map("${:.2f}".format)

popular_item =popular_item.sort_values("Purchase Count", ascending= False)
popular_item.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
#Most popular item sort
most_popular_item =popular_item.sort_values("Total Purchase Value", ascending= False)
most_popular_item["Total Purchase Value"] = most_popular_item['Total Purchase Value'].map("${:.2f}".format)
most_popular_item.head()
```
